```python
def get_date(date_set):
  for i in date_set:
    date = i
    return date
```

```python
def get_word_rows_from_utt (df):
#takes in a dataframe created from: conv1 = sbc_1.utterances(), df = pd.DataFrame(conv1)
  #dfnew =pd.DataFrame(df.drop(columns=["time_marks"]))
  dfnew = df.explode("tokens")
  dfnew = pd.concat([dfnew.drop(['tokens'], axis=1), dfnew['tokens'].apply(pd.Series)], axis=1)
  return dfnew
```

```python
def get_part_ages (reader):
  parts = reader.participants()
  ages = []
  for f in parts: 
    x = reader.ages(f)[0][0]
    ages.append(x)
  ad = pd.DataFrame()
  ad['participant'] = list(parts)
  ad['age'] = ages
  return ad
```

```python
def get_words_per_part (reader, df):
  parts = reader.participants()
  totals = []
  for f in parts: 
    try:
      x = df.participant.value_counts()[f]
      totals.append(x)
    except:
      print(f)
      totals.append(0)
  ad = pd.DataFrame()
  ad['participant'] = list(parts)
  ad['total'] = totals
  df = df.merge(ad.set_index('participant'), on='participant', validate='m:1')
  return df
  
```

```python
def get_likes_per_part (reader, df):
  parts = reader.participants()
  totals = []
  for f in parts: 
    try:
      x = df.participant.value_counts()[f]
      totals.append(x)
    except:
      print(f)
      totals.append(0)
  ad = pd.DataFrame()
  ad['participant'] = list(parts)
  ad['likes'] = totals
  df = df.merge(ad.set_index('participant'), on='participant', validate='m:1')
  return df
```

```python
def conv_to_df(filepath):
    
    # import from specific file
    reader = pla.read_chat(filepath)
    
    # get the utterances as a list of tuples
    conv = reader.utterances()
    
    # put this list of tuples into a DataFrame
    df = pd.DataFrame(conv)

    ages = get_part_ages(reader)
    df = df.merge(ages.set_index('participant'), on='participant', validate='m:1')

    df['Utterance_ID'] = df.index

    df = get_word_rows_from_utt(df)

    df = get_words_per_part(reader, df)

    dt1 = df.loc[df['word'].str.contains('like')]
    dt2 = df.loc[df['word'].str.contains('Like')]
    df = dt1.append(dt2)

    df = get_likes_per_part(reader, df)

    dd= reader.dates_of_recording()
    df['date'] = get_date(dd)

    df['Conv_ID'] = os.path.basename(filepath).rstrip('.cha')

    df = df[['date','Conv_ID', 'Utterance_ID', 'participant','age','word','pos','tiers','total','likes']]
    
    return df
```
